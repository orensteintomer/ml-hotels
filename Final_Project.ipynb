###### a. Find out for each (Snapshot Date+ Checkin Date + DayDiff + Hotel Name + WeekDay) the best Discount Code(max discount) and save to new CSV file

We will use group by on the wanted fields, and in order to find the best dicount code we will use MAX function on DiscountPerc field

```python
# We will firstly group by wanted fields
gbdata = dffulldata.groupby(['Snapshot Date', 'Checkin Date', 'DayDiff', 'Hotel Name', 'WeekDay'])

# Now lets see the grouped by data
gbdata.size()
```

```python
# This function creates one row which represents the whole group
# The function finds each group's best discount code
# Input: group
# Output: row of ['Snapshot Date', 'Checkin Date', 'DayDiff', 'Hotel Name', 'WeekDay', 'Discount Code']
def getGroupWithDiscountCode(group):

    # Finds group's max discount precent and gets its row index
    rowId = group['DiscountPerc'].idxmax(axis=1)
    
    # Gets the discount code from the row with the max discount precent
    bestCode = group.loc[rowId]['Discount Code']

    # Takes random row from the group for creating the group's representive row
    randomRow = group.iloc[0]

    return {'Snapshot Date': randomRow['Snapshot Date'],
            'Checkin Date': randomRow['Checkin Date'],
            'DayDiff': randomRow['DayDiff'],
            'Hotel Name': randomRow['Hotel Name'],
            'WeekDay': randomRow['WeekDay'],
            'Discount Code': bestCode }

# Gets list of rows (one for each group) - with its best discount code
datalist = gbdata.apply(getGroupWithDiscountCode)

# Creates new dataframe for the results
finaldf = pd.concat([pd.DataFrame([row], columns=['Snapshot Date',
                                                  'Checkin Date',
                                                  'DayDiff',
                                                  'Hotel Name',
                                                  'WeekDay',
                                                  'Discount Code']) for index, row in datalist.iteritems()], ignore_index = True)
```

###### Naive bayes

```python
# Lets calculate the TP and FP for each discount code
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(nbexpected, nbpredicted)

FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)  
FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
TP = np.diag(confusion_matrix)
TN = confusion_matrix.sum() - (FP + FN + TP)

TPR = TP / (TP+FN)
FNR = FN / (TP+FN)
FPR = FP / (FP+TN)
print('False Negative')
print(FNR)
print('False Poisitive')
print(FPR)
```

```python
# Lets calculate the ROC for each discount code
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

exp = label_binarize(nbexpected.as_matrix(), classes=[1, 2, 3, 4])
pred = label_binarize(nbpredicted, classes=[1, 2, 3, 4])

n_classes = 4

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(exp[:, i], pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    
for i in range(n_classes):
    plt.figure()
    plt.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f)' % roc_auc[i])
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic ' + 'discount code ' + str(i+1))
    plt.legend(loc="lower right")
    plt.show()
```

###### a. find the 150 hotels which have the most rows in our origin csv and take only their data.


```python
# loads origin csv data
df = pd.read_csv("hotels_data.csv")
df.head()

# Group by hotel name
gbhtlname = df.groupby(['Hotel Name'])

# Sort by count and take first 150
hotelstocountdf = gbhtlname.agg(['count'])['Days'].sort_values(by=['count'], ascending=False)
hotelstocountdf = hotelstocountdf.head(150)

# Take only that hotels' data
df = df.loc[df['Hotel Name'].isin(hotelstocountdf.index)]
df
```

###### b. find the 40 check in dates which have the most rows and take only their data.

```python
# Group by check in date
gbcheckin = df.groupby(['Checkin Date'])

# Sort by count and take first 40
datestocountdf = gbcheckin.agg(['count'])['Days'].sort_values(by=['count'], ascending=False)
datestocountdf = datestocountdf.head(40)

# Take only that Checkin Date' data
df = df.loc[df['Checkin Date'].isin(datestocountdf.index)]
df

gbhtlname = df.groupby(['Hotel Name'])
gbhtlname.size()
```

###### c. for each check in date - take the lowest price(after discount) for each discount code.(4 codes) --> each hotel will have 160 columns of prices(40 dates * 4 discount codes)

```python
# Extracts dates list 
dates = datestocountdf.index.values

# Map the dates list into list of lists - from each date in origin list creates list of date + discount code
l = list(map(lambda x: list(map(lambda y: x + " " + str(y),[1,2,3,4])),dates))

# Flat the list
flat_list = [item for sublist in l for item in sublist]

# Creates dataframe for results - add column for each checkin date + discount code
dfresults = pd.DataFrame(columns=['Hotel Name'] + flat_list)

# Group by hotel name
gbhtlname = df.groupby(['Hotel Name'])

# Iterate over hotels groups
for htlname, htlgroup in gbhtlname:
    
    # Creates current hotel row
    htlrow = { 'Hotel Name': htlname }
    
    # Group by check in date
    gbcheckin = htlgroup.groupby(['Checkin Date'])
    
    # Iterate over check in dates in current hotel
    for chkdname, chkdgroup in gbcheckin:
        
        # Group by discount
        gbdiscode = htlgroup.groupby(['Discount Code'])
        
        # Iterate over discount codes in current check in date in current hotel
        for discodename, discodegroup in gbdiscode:
            
            # Find row of minimum value of Discount Price 
            rowId = discodegroup['Discount Price'].idxmin(axis=1)
            
            # Gets the Discount Price
            minprice = discodegroup.loc[rowId]['Discount Price']
            
            # Sets current checkin + discount code value
            htlrow[chkdname + " " + str(discodename)] = minprice
            
    # Append current hotel row to results data frame       
    dfresults = dfresults.append(htlrow, ignore_index=True)

# Assign missing values (missing dates for some hotels) to -1 default value AS REQUESTED
dfresults = dfresults.fillna(-1)
dfresults
```

###### d. normalize each price from 1 to 100.

```python
import math

# Normaliztion formula
def normalform(maximum, minimum, value):
    if (maximum == minimum):
        return 100
    else:
        return ((value-minimum)/(maximum-minimum)) * 100
    
dfvalues = dfresults.drop('Hotel Name', axis=1)

# Runs over all results
for rindex, row in dfvalues.iterrows():
    
    # Resests max and min
    maxval = row.max()
    minval = maxval
    
    # Runs over all columns and find minimum value except of -1
    for column in dfvalues:
        
        # Checks if current cell's value is different than -1 and smaller than current minimum value
        if row[column] != -1 and row[column] < minval:
            minval = row[column]
    
    # Normalize current row's cells
    dfresults.iloc[rindex,1:161] = dfresults.iloc[rindex,1:161].apply(lambda x: -1 if x == -1 else normalform(maxval, minval, x))
    
dfresults      
```

##### b. we will use our ML library scikit only for creating the model, and combine scipy library only for plotting the Dendrogram (script from scikit github repository - not offical) - works not so good..

```python
import numpy as np

from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

def plot_dendrogram(model, **kwargs):

    # Children of hierarchical clustering
    children = model.children_

    # Distances between each pair of children
    # Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

    # The number of observations contained in each cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

    # Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
from sklearn.cluster import AgglomerativeClustering

# Drops hotel name 
dfvalues = dfresults.drop('Hotel Name', axis=1)

# Fit the clustering model
model = AgglomerativeClustering(linkage='ward').fit(dfvalues)

plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model, labels=model.labels_)
plt.show()
```

```python
# Change columns to have numeric values for later ML algorithms usage
from pyspark.sql.functions import udf,unix_timestamp
from pyspark.sql.types import IntegerType

def daytoint(x):
    return {
        'Sunday': 1,
        'Monday': 2,
        'Tuesday': 3,
        'Wednesday': 4,
        'Thursday': 5,
        'Friday': 6,
        'Saturday': 7
    }[x]

daytoint_udf = udf(daytoint, IntegerType())

df = df.withColumn('Snapshot Date', unix_timestamp(col('Snapshot Date'))). \
        withColumn('Checkin Date', unix_timestamp(col('Checkin Date'))). \
        withColumn('WeekDay', daytoint_udf(col('WeekDay')))
```

```python
# Now we will handle the hotels name column - use StringIndexer
from pyspark.ml.feature import StringIndexer
indexer = StringIndexer(inputCol="Hotel Name", outputCol="Hotel Name_idx")
indexed = indexer.fit(df).transform(df)
indexed.head(5)
```

##### It is time to prepare the data in MLLIB structure - features and class columns

```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

assembler = VectorAssembler(
  inputCols=['Snapshot Date',
             'Checkin Date',
             'Hotel Name_idx',
             'DayDiff',
             'WeekDay'],
    outputCol="features")

output = assembler.transform(indexed)
final_data = output.select("features", "Discount Code")

# Split the data into train and test
train_data,test_data = final_data.randomSplit([0.75,0.25])
```

###### Naive bayes

```python
from pyspark.ml.classification import NaiveBayes

# Creates the trainer and set its parameters
nb = NaiveBayes(modelType="multinomial", labelCol="Discount Code")

# Trains the model
nbmodel = nb.fit(train_data)

# Select example rows to display
nbpredictions = nbmodel.transform(test_data)
nbpredictions.show()
```

```python
# Computes accuracy on the test set
nbevaluator = MulticlassClassificationEvaluator(labelCol="Discount Code", predictionCol="prediction",
                                              metricName="accuracy")
accuracy = nbevaluator.evaluate(nbpredictions)
print("Naive bayes test set accuracy = " + str(accuracy))
```

###### decision tree 

```python
from pyspark.ml.classification import DecisionTreeClassifier

# Creates the trainer and set its parameters
dt = DecisionTreeClassifier(labelCol="Discount Code", featuresCol="features", maxBins=554)

# Trains the model
dtmodel = dt.fit(train_data)

# Select example rows to display
dtpredictions = dtmodel.transform(test_data)
dtpredictions.show()
```

```python
# Computes accuracy on the test set
dtevaluator = MulticlassClassificationEvaluator(labelCol="Discount Code", predictionCol="prediction",
                                              metricName="accuracy")
accuracy = dtevaluator.evaluate(dtpredictions)
print("Decision Tree test set accuracy = " + str(accuracy))
```

```python
from pyspark.mllib.evaluation import MulticlassMetrics

nbpredictionAndLabels = nbpredictions.rdd.map(lambda x: (x["prediction"], float(x["Discount Code"])))
metrics = MulticlassMetrics(nbpredictionAndLabels)

print('True Positive')
print(metrics.truePositiveRate(1.0))
print(metrics.truePositiveRate(2.0))
print(metrics.truePositiveRate(3.0))
print(metrics.truePositiveRate(4.0))
print('False Poisitive')
print(metrics.falsePositiveRate(1.0))
print(metrics.falsePositiveRate(2.0))
print(metrics.falsePositiveRate(3.0))
print(metrics.falsePositiveRate(4.0))
```

```python
## ROC

from pyspark.mllib.evaluation import BinaryClassificationMetrics

def binarize(predAndLabel, index):
    
    first = 0.0
    second = 0.0
    if predAndLabel[0] == index:
        first = 1.0
    if predAndLabel[1] == index:
        second = 1.0
    return (first, second)

for index in range(1,5):
    
    curPredictionAndLabels = nbpredictionAndLabels.map(lambda x: binarize(x, index))

    # Instantiate metrics object
    metrics = BinaryClassificationMetrics(curPredictionAndLabels)

    # ROC
    print("ROC For Discount code " + str(index) + " is " + str(metrics.areaUnderROC))
```
